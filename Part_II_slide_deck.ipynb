## Aperçu de l'enquête
> Notre enquête qui à porté sur l'exploration de l'ensemble des données **prosperLoanData** nous a permi d'observer la distribution du statut d'un prêt. Par cela, nous avons également pu dégager les correlations entre le statut du prêt et les autres caractéristiques telles que la mensualité prévu de paiement, le montant initial du prêt, le statut de l'emploi de l'emprunteur, la durée du prêt ainsi que la profession de l'emprunteur.
## Aperçu de l'ensemble de données
> L'ensemble de données **prosperLoanData** dont nous sommes sur le point d'explorer contient 113 937 prêts avec 81 variables sur chaque prêt, y compris le montant du prêt, le taux de l'emprunteur (ou taux d'intérêt), le statut actuel du prêt, le revenu de l'emprunteur, et bien d'autres.

```python
# import des packages et définition des tracés à intégrer en ligne
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppression des avertissements de la sortie finale
import warnings
warnings.simplefilter("ignore")
```

```python
# chargement de l'ensemble de données dans une base de données pandas
df = pd.read_csv('prosperLoanData.csv')
```

## **Distribution des statuts des prêt**

> Nous présentons ici la distribution des statuts des prêt qui, comme on peut le voir, est largement dominé pas les prêts *En cours*. 
> L'interêt pour cette présentation pour nous est ressortir les différentes proportions de l'état des prêts.

```python
sorted_counts = df['LoanStatus'].value_counts()
plt.figure(figsize=[8,6])
plt.pie(sorted_counts, labels = sorted_counts.index, startangle = 90, counterclock = False)
plt.axis('square')
plt.title('Répartition des statuts des prêts')
plt.show()
```

## **Répartition  des statuts d'emploi des emprunteurs**

> La répartition des types d'emploi est largement dominée par les emprunteurs *Employé*

```python
ordre = df.EmploymentStatus.value_counts().index
plt.figure(figsize=[10,5])
sb.countplot(x=df['EmploymentStatus'], color= 'blue', order=ordre)
plt.title("Répartition  des types d'emploi des emprunteurs")
plt.xlabel("Types d'emploi")
plt.ylabel('Nombre')
plt.xticks(rotation=45)
plt.show()
```

## **Correlation entre le montant initial et la mensualité prévue du prêt**
> Le montant initial et la mensualité prévue du prêt suivent de façon générale une correlation positive relativement faible.


```python
plt.figure(figsize = [8, 5])
plt.scatter(data = df, x = 'LoanOriginalAmount', y = 'MonthlyLoanPayment')
plt.xlabel("montant initial")
plt.ylabel("mensualité du prêt")
plt.title("Correlation entre le montant initial et la mensualité prévue du prêt")
plt.show()
```

## **Correlation entre mensualité prévue et statut du prêt**
> Mensualité prévue du prêt élevée est dominée par les prêts *En cours*; par contre, les prêts à faible mensualité ont été annulé.

```python
plt.figure(figsize = [8, 6])
base_color = sb.color_palette()[0]
sb.barplot(data=df, x='MonthlyLoanPayment',y='LoanStatus', color=base_color)
plt.title("Correlation entre mensualité prévue et statut du prêt")
plt.xlabel("Mensualité prévue du prêt")
plt.ylabel("statut du prêt")
plt.show()
```

## **Relation entre le statut de l'emploi, mensualité de paiement et état du prêt**
> Les prêts annulés ont une mensualité prévue de moins de 100$ et les emplois impactés sont ceux non disponible et à temp plein.

```python
plt.figure(figsize = [10, 8])
sb.barplot(x='LoanStatus', y='MonthlyLoanPayment', data=df, palette='bright', hue='EmploymentStatus')
plt.title("Relation entre le statut de l'emploi, mensualité de paiement et état du prêt")
plt.ylabel("Mensualité prévue du prêt")
plt.xlabel("statut du prêt")
plt.xticks(rotation=90)
plt.show()
```
